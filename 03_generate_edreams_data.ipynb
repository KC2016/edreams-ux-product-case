```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

n_records = 1000
```

```python
# 1. Generate data for Focus Area 1: Seat Map & Dynamic Pricing
seats = ['window', 'aisle', 'middle']
seat_types = np.random.choice(seats, size=n_records, p=[0.4,0.4,0.2])
flight_occupancy = np.random.randint(50,100, size=n_records) # Flight load factor in %

base_price = 50
seat_prices = []
```

```python
# seats
# seat_types
# flight_occupancy
```

```python
# Apply dynamic pricing rules based on seat type and flight demand
for s_type, occupancy in zip(seat_types, flight_occupancy):
    price = base_price
    if s_type in ['window', 'aisle']:
        price += 15 + (occupancy * 0.2) # Premium charge for high demand
    else:
        price += 5 - (occupancy * .01)  # Discount to offload middle seats if flight is empty
    seat_prices.append(round(price, 2))
        
```

```python
# seat_prices
```

```python
# 2. Generate data for Focus Area 2: Accessibility & Ancillary Upsell (Baggage)
is_pcd = np.random.choice([0,1], size=n_records, p=[0.92, 0.08]) # 1 = Visually impaired user
layout_bug_active = np.random.choice([0,1], size=n_records, p=[0.5,0.5]) # Before vs. After fix
```

```python
# is_pcd
# layout_bug_active
```

```python
baggage_upsell=[]
for pcd, bug in zip(is_pcd, layout_bug_active):
    if pcd == 1 and bug == 1:
        # Bug active tanked conversion for visually impaired users (5% conversion)
        baggage_upsell.append(np.random.choice([0,1], p=[0.95, 0.05]))
    elif pcd == 1 and bug == 0:
        # Post-fix recovery/normalization (40% conversion)
        baggage_upsell.append(np.random.choice([0,1], p=[0.60, 0.40]))
    else:
        # Standard users baseline (35% conversion)
        baggage_upsell.append(np.random.choice([0,1], p=[0.65,0.35]))
```

```python
# baggage_upsell
```

```python
# 3. Generate data for Focus Area 3: Risk Mitigation & Postal API Impact
uses_postal_api = np.random.choice([0,1], size=n_records, p=[0.4,0.6]) # 1 = Address auto-validated
payment_declined = []
```

```python
# Simulate how address validation reduces checkout payment declines
for api in uses_postal_api:
    if api == 1:
        payment_declined.append(np.random.choice([0,1], p=[0.97, 0.03])) # Only 3% decline rate
    else:
        payment_declined.append(np.random.choice([0, 1], p=[0.85, 0.15])) # 15% decline rate due to typos
```

```python
# payment_declined
```

```python
# --- COMBINE ALL FEATURES INTO A SINGLE INTEGRATED DATAFRAME ---
df_edreams = pd.DataFrame({
    'transaction_id': range(1000, 1000 + n_records),
    'route': np.random.choice(['bcn-ory', 'bcn-cdg'], size=n_records), # Lowercase routes
    'seat_type': seat_types,
    'flight_occupancy_pct': flight_occupancy,
    'seat_price_paid': seat_prices,
    
    # --- Focus Area 2: Standby Legacy Columns ---
    'is_visually_impaired': is_pcd,
    'layout_bug_active': layout_bug_active,
    'baggage_upsell_purchased': baggage_upsell,

#     # --- Focus Area 2: New Localization UX Columns ---
#     'user_country': user_countries,
#     'user_language': user_languages,
#     'currency_preference': currencies,
#     'app_forced_language': app_forced_languages,
#     'session_abandoned': session_abandoned,
  
    # --- Focus Area 3: Risk Mitigation & Payment ---
    'uses_postal_api': uses_postal_api,
    'payment_declined': payment_declined
})
```

### ✈️ Domain Context: Airport Route Selection & Industry Standards

In this dataset simulation, the routes `bcn-ory` and `bcn-cdg` represent real-world commercial flights departing from Barcelona El Prat Airport (BCN) and arriving at the two primary airports serving Paris, France. Including both destinations allows the AI Agent to perform precise customer segmentation based on airline business models (Low-Cost vs. Legacy).

* **ory (Paris-Orly Airport):** Located south of Paris, Orly is a major hub for secondary European traffic, domestic flights, and low-cost carriers (e.g., Vueling, Transavia). It structurally represents budget-conscious travelers and short-haul business commuters.
* **cdg (Paris-Charles de Gaulle Airport):** Located north of Paris, Charles de Gaulle is the largest international airport in France and a primary global hub for legacy carriers (e.g., Air France) and long-haul intercontinental connections. It represents premium travelers and connecting passengers.
